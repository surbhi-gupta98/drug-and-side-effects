```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('/content/drugs_side_effects_drugs_com.csv')
```

```python
data.head()
```

```python
data.isnull().sum(0)
```

```python
columns_to_replace = ['side_effects', 'generic_name', 'drug_classes', 'brand_names','rx_otc','pregnancy_category']
data[columns_to_replace] = data[columns_to_replace].replace(np.NaN, 'unknown')
```

```python
data.drop('rating',axis=1,inplace=True)
```

```python
columns_to_drop = ['related_drugs','medical_condition_description','no_of_reviews','drug_link','medical_condition_url']
data.drop(columns_to_drop,axis=1,inplace=True)
```

```python
data.isnull().sum()
```

```python
data['alcohol'].replace('X',1,inplace=True)
```

```python
data['alcohol'].fillna(0,inplace=True)
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
data.info()
```

```python
data['alcohol'] = data['alcohol'].astype(int)
```

```python
data.head()
```

```python
data.info()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.shape
```

```python
data.to_csv('drugs and side effects.csv',index=False)
```
